```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
month=np.array(['E','F','M','A','Ma'])
sales=np.array([20,25,30,28,35])
## configurar grafico
plt.figure(figsize=(6, 8))
##crear el grafico
plt.plot(month,sales,marker='X',color='blue')
plt.title('ventas mensuales por producrtos')
plt.xlabel('meses')
plt.ylabel('ventas en miles')
plt.show()
```

```python

##configure nuevo grafico
plt.figure(figsize=(6,8))
hour=[2,3,4,5,6,7,8]
exan=[22,34,56,67,87,90,99]

plt.scatter(hour,exan,color='green')
plt.xlabel('Hour')
plt.ylabel('Score')
plt.show()
```

```python
## personalizacion de graficos

plt.figure(figsize=(6,8))
hour=[2,3,4,5,6,7,8]
exan_1=[22,34,56,67,87,90,99]
exan_2=[32,44,55,69,71,79,86]
##crar grafico de dispercion 
plt.scatter(hour,exan_1,marker='x',color='green',linestyle='-', linewidths=2,label='student_1')
plt.scatter(hour,exan_2,marker='o',color='blue',linestyle='--',linewidths=2,label='student_2')


plt.xlabel('Hour')
plt.ylabel('Score')
plt.show()
```

```python
##graficos de barras y paid
plt.figure(figsize=(8,9))
categories=['producto a','producto b','producto c']
sales=[120,150,78]
plt.bar(categories,sales,color='blue',label='ventas mensaules')

plt.title('ventas de productos en un mes')
plt.xlabel('productos')
plt.ylabel('ventas')
plt.legend()
plt.show()
```

```python
##graficos de barras y paid
plt.figure(figsize=(10,10))
categories=['producto a','producto b','producto c']
sales=[120,150,78]
plt.bar(categories,sales,color='blue',label='ventas mensaules')
plt.annotate('maximo de ventas',xy=('producto b',150),xytext=('producto c',160),
             arrowprops=dict(facecolor='black',shrink=0.05))
plt.title('ventas de productos en un mes')
plt.xlabel('productos')
plt.ylabel('ventas')
#plt.legend()
plt.show()
```

```python
##graficos de barras y paid
plt.figure(figsize=(8,9))
categories=['producto a','producto b','producto c']
sales=[120,150,78]
plt.barh(categories,sales,color='blue',label='ventas mensaules')

plt.title('ventas de productos en un mes')
plt.ylabel('productos')
plt.xlabel('ventas')
plt.show()
```

```python
# Crear un diagrama de pastel (pie chart) utilizando las variables existentes

# Configurar el tamaño de la figura
plt.figure(figsize=(8, 8))

# Crear el diagrama de pastel
# 'categories' contiene las etiquetas de cada porción del pastel
# 'sales' contiene los valores correspondientes a cada categoría
plt.pie(sales, labels=categories, autopct='%1.1f%%', startangle=140)

# Añadir un título al gráfico
plt.title('Distribución de ventas por producto')

# Mostrar el gráfico
plt.show()
```

```python
plt.figure(figsize=(8,9))

products=['Muebles','Artefactos','Videojuegos','Libros']
sales=[50,71,30,25]

plt.pie(sales,labels=products,autopct='%1.1f%%' ,startangle=140,colors=['blue','orange','lightcoral','black'])

plt.title('Participacion de ventas por categorias')
plt.axis('equal')
plt.show()
```

Histogramas

```python
## generando data
data=np.random.normal(170,10,200)
print(data)
```

```python
plt.hist(data,color='salmon',bins=55,edgecolor='black',alpha=0.7)
plt.title('Distribucion de altura')
plt.xlabel('Altura')
plt.ylabel('Dencidad')
plt.show()
```

Boxplot 

```python
np.random.seed(0)
ages=[np.random.normal(30,5,100),
      np.random.normal(40,5,100),
      np.random.normal(35,5,100),]
```

```python
plt.boxplot(ages,patch_artist=True,notch=True,vert=True,
            labels=['Grupo1','Grupo2','Grupo3'])

plt.title('Distribucion de edades')
plt.xlabel('Grupo')
plt.ylabel('Edad Años')
plt.show()
```
